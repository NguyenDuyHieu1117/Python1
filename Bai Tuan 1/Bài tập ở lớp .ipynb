```python
# Bai 15
class SinhVien:
    def __init__(self, maSV, ten, namSinh, diemTB):
        self.maSV = maSV
        self.ten = ten
        self.namSinh = namSinh
        self.diemTB = diemTB


# Cho một mảng có n sinh viên
n = int(input("Nhap so sinh vien: "))
ds = []

for i in range(n):
    print("Nhap sinh vien:", i+1)

    maSV = input("Mã SV: ")
    ten = input("Tên: ")
    namSinh = int(input("Năm sinh: "))
    diemTB = float(input("Điểm TB: "))

    sv = SinhVien(maSV, ten, namSinh, diemTB)
    ds.append(sv)


# c. Sinh viên lên lớp
def len_lop(ds):
    dem = 0
    for sv in ds:
        if sv.diemTB >= 5:
            dem += 1
    return dem

print("Số SV đủ điều kiện lên lớp:", len_lop(ds))


# d. Sinh viên đủ 20 tuổi
from datetime import datetime
nam_hien_tai = datetime.now().year

print("Sinh viên đủ 20 tuổi:")
for sv in ds:
    if nam_hien_tai - sv.namSinh >= 20:
        print(sv.maSV, sv.ten, sv.namSinh, sv.diemTB)


# e. Sinh viên hệ DH
def dem_DH(ds):
    dem = 0
    for sv in ds:
        if sv.maSV[2:4] == "DH":
            dem += 1
    return dem

print("Số sinh viên hệ đại học:", dem_DH(ds))


# f. Sinh viên tên Lan
def dem_Lan(ds):
    dem = 0
    for sv in ds:
        if "Lan" in sv.ten:
            dem += 1
    return dem

print("Số sinh viên tên Lan:", dem_Lan(ds))


# g. Sinh viên họ Phan
def dem_Phan(ds):
    dem = 0
    for sv in ds:
        if sv.ten.startswith("Phan"):
            dem += 1
    return dem

print("Số sinh viên họ Phan:", dem_Phan(ds))
```
