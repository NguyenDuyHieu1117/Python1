```python
# Bai 1
def tron_mang(a, b):
    c = []
    n = min(len(a), len(b))
    # Lấy min của 2 mảng
    for i in range(n):
        c.append(min(a[i], b[i]))
    # Thêm phần tử còn lại
    if len(a) > len(b):
        c.extend(a[n:])
    else:
        c.extend(b[n:])

    return c
# Nhập mảng a
a = list(map(int, input("Nhập mảng a: ").split()))
# Nhập mảng b
b = list(map(int, input("Nhập mảng b: ").split()))
# Trộn mảng
ket_qua = tron_mang(a, b)
print("Mảng kết quả:", ket_qua)
```

```python
# Bai 2
# Nhập ma trận
def nhap_ma_tran(m, n):
    a = []
    for i in range(m):
        hang = list(map(int, input(f"Nhập dòng {i+1}: ").split()))
        a.append(hang)
    return a
# Xuất ma trận
def xuat_ma_tran(a):
    for hang in a:
        print(*hang)
# a. Tổng tam giác trên đường chéo phụ
def tong_tam_giac_tren_cheo_phu(a, n):
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong
# b. Đổi số âm thành trị tuyệt đối
def doi_am_thanh_duong(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])
# c. Thay số chẵn bằng x
def thay_so_chan(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x
# d. Kiểm tra toàn chẵn
def kiem_tra_toan_chan(a):
    for hang in a:
        for x in hang:
            if x % 2 != 0:
                return False
    return True
# e. Kiểm tra ma trận đối xứng
def doi_xung(a, n):
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True
# f. Đường chéo chính tăng dần
def cheo_chinh_tang(a, n):
    for i in range(n-1):
        if a[i][i] > a[i+1][i+1]:
            return False
    return True

# g. Xuất tam giác dưới đường chéo phụ
def xuat_tam_giac_duoi_cheo_phu(a, n):
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()
# h. Đường chéo phụ giảm dần
def cheo_phu_giam(a, n):
    for i in range(n-1):
        if a[i][n-1-i] < a[i+1][n-2-i]:
            return False
    return True
m = int(input("Nhập số dòng m: "))
n = int(input("Nhập số cột n: "))
a = nhap_ma_tran(m, n)
print("\nMa trận ban đầu:")
xuat_ma_tran(a)
# a
print("\nTổng tam giác trên đường chéo phụ:", tong_tam_giac_tren_cheo_phu(a, n))
# b
doi_am_thanh_duong(a)
print("\nMa trận sau khi đổi số âm:")
xuat_ma_tran(a)
# c
x = int(input("\nNhập số thay thế x: "))
thay_so_chan(a, x)
print("\nMa trận sau khi thay số chẵn:")
xuat_ma_tran(a)
# d
print("\nMa trận toàn chẵn:", kiem_tra_toan_chan(a))
# e
print("Ma trận đối xứng:", doi_xung(a, n))
# f
print("Đường chéo chính tăng dần:", cheo_chinh_tang(a, n))
# g
print("\nTam giác dưới đường chéo phụ:")
xuat_tam_giac_duoi_cheo_phu(a, n)
# h
print("\nĐường chéo phụ giảm dần:", cheo_phu_giam(a, n))
```
